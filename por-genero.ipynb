```python
import pandas
import matplotlib

# para que los gráficos se muestren en el documento
%matplotlib inline

# URL del CSV con los datos
file = 'http://opendata.ugr.es/dataset/a6353c6f-130a-48bb-82a7-500a2fc457e1/resource/657ca05e-3c40-4121-8c1e-b398e1cf30bf/download/201516matriculasgradoramas.csv'

# variables para leer el CSV
encoding="ISO-8859-1"
sep=','
skipinitialspace=True

```

```python
df_original = pandas.read_csv(file, encoding= encoding, sep= sep, skipinitialspace= skipinitialspace)
```

```python
df_original.head()
```

```python
# Nos quedamos sólo con los datos intersantes
df_limpio = df_original.iloc[:,1:4]
```

```python
df_limpio.head()
```

```python
# gráfico de barras horizontales
df_limpio.plot(kind='barh', stacked=True, x=0 ,figsize= (10,20))
```
